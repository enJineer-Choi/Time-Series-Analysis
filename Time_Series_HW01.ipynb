##5


```r
b=scan("book.txt")
head(b)
```

```r
t5=1:length(b)
plot.ts(b)  ##plot.ts를 활용해 시계열 그림을 그림
```

- (2)
위 시계열 그림을 통해 추세성분이 존재한다고 할 수 있다

- (3)

```r
df5=data.frame(
    t=1:length(b),
    b=b
)
df5
```

```r
m5=lm(b~t5,data=df5)
m5
```

```r
summary(m5)
```

```r
new_dt_5=data.frame(
    t=31:42,
    y=as.factor(1:12)
)
```

```r
predict_result_5=as.data.frame(predict(m5,newdata=new_dt_5,interval="prediction",level = 0.95))
predict_result_5
```

```r
plot(e~d,dt,type = 'l')
```

```r
dt$y=as.factor(as.integer(cycle(dep)))
dt$t=1:length(e)
```
