```r
z=ts(matrix(rnorm(300),100,3),start=c(2000),frequency=12)
plot(z)
```

```r
plot.ts(z) ##plot.ts()는 따로 그림
```

```r
ts.plot(z) ##ts.plot()는 겹쳐서 그림
```

```r
plot.ts(z,plot.type="single",col=1:3,lty=1:3)
```

```r
ts.plot(z,col=1:3) ##시간이 달라도 알아서 그려줌.
```
